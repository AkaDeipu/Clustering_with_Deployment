## Outlier Counts per Feature

```python
outlier_counts = {}
for col in df2.columns:
  q1 = df2[col].quantile(0.25)
  q3 = df2[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df2[(df2[col]<lower_bound) | (df2[col]> upper_bound)]
  outlier_counts[col] = len(outliers)
for col, count in outlier_counts.items():
  print(f"{col}: {count}")
```

## K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
```

```python
wcss = []
silhouette = []
calinski = []
davies = []
```

```python
for k in range(2,12):
  kmeans = KMeans(n_clusters = k, random_state = 42, init = "k-means++")
  labels = kmeans.fit(df3)

  wcss.append(kmeans.inertia_)
  silhouette.append(silhouette_score(df3, labels.labels_))
  calinski.append(calinski_harabasz_score(df3, labels.labels_))
  davies.append(davies_bouldin_score(df3, labels.labels_))
```

```python
labels.labels_ # for each row a cluster is assigned
```

```python
silhouette
```

1. WCSS (Within-Cluster Sum of Squares / Inertia)
What it measures: How tightly the points are grouped around their cluster centers.

Interpretation:

Lower WCSS â†’ clusters are more compact.

As you increase k, WCSS always decreases (more clusters = smaller distances).

You usually look for the â€œelbow pointâ€ where adding more clusters doesnâ€™t reduce WCSS much.

2. Silhouette Score
Range: -1 to +1

What it measures: How similar a point is to its own cluster compared to other clusters.

Interpretation:

Close to +1 â†’ well-separated, dense clusters.

Around 0 â†’ overlapping clusters.

Negative â†’ points may be in the wrong cluster.

Use: A higher silhouette score means better-defined clusters.

3. Calinskiâ€“Harabasz Index
Formula: Ratio of between-cluster dispersion to within-cluster dispersion.

Interpretation:

Higher values â†’ better separation between clusters.

It rewards compact, well-separated clusters.

Use: Good for comparing different k values; the higher, the better.

4. Daviesâ€“Bouldin Index
Range:
0
 to âˆž.

What it measures: Average similarity between each cluster and its most similar other cluster.

Interpretation:

Lower values â†’ better clustering (clusters are distinct and far apart).

Higher values â†’ clusters are overlapping or poorly separated.

## Hierarchical Clustering

### Dendogram

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import itertools
```

```python
methods = ['ward', 'single', 'complete', 'centroid','average']
metrics = ['euclidean', 'cityblock', 'cosine']

for method, metric in itertools.product(methods, metrics):
  try:
    if method == 'ward' and metric != 'euclidean':
      continue
    plt.figure(figsize = (8,6))
    hc = linkage(df3, method = method, metric = metric)
    dendrogram(hc, truncate_mode = 'level', p = 5)
    plt.title(f"Dendrogram (Method: {method}, Metric: {metric})")
    plt.xlabel('Samples')
    plt.ylabel('Distance')
    plt.show()
  except Exception as e:
    print(f"Skipped method = {method}, metric = {metric} due to error {e}")
```

### Evaluation of Hierarchical

```python
methods = ['ward', 'single', 'complete', 'average']
metrics = ['euclidean', 'cityblock', 'cosine']

results = []

for method, metric in itertools.product(methods, metrics):
  try:
    if method == 'ward' and metric != 'euclidean':
      continue

    z = linkage(df3, method = method, metric =metric)

    hc_labels = fcluster(z, criterion = 'maxclust', t = 5)

    hc_silhouette = silhouette_score(df3, hc_labels)
    hc_calinski = calinski_harabasz_score(df3, hc_labels)
    hc_davies = davies_bouldin_score(df3, hc_labels)

    results.append({
        'Method': method,
        'Metric': metric,
        'Silhouette': hc_silhouette,
        'Calinski': hc_calinski,
        'Davies': hc_davies
    })

  except Exception as e:
    print(f"Skipped method = {method}, metric = {metric} due to error {e}")
```

## Hierarchical Clustering

### Dendogram

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import itertools
```

```python
methods = ['ward', 'single', 'complete', 'centroid','average']
metrics = ['euclidean', 'cityblock', 'cosine']

for method, metric in itertools.product(methods, metrics):
  try:
    if method == 'ward' and metric != 'euclidean':
      continue
    plt.figure(figsize = (8,6))
    hc = linkage(df6, method = method, metric = metric)
    dendrogram(hc, truncate_mode = 'level', p = 5)
    plt.title(f"Dendrogram (Method: {method}, Metric: {metric})")
    plt.xlabel('Samples')
    plt.ylabel('Distance')
    plt.show()
  except Exception as e:
    print(f"Skipped method = {method}, metric = {metric} due to error {e}")
```

### Evaluation of Hierarchical

```python
methods = ['ward', 'single', 'complete', 'average']
metrics = ['euclidean', 'cityblock', 'cosine']

results = []

for method, metric in itertools.product(methods, metrics):
  try:
    if method == 'ward' and metric != 'euclidean':
      continue

    z = linkage(df6, method = method, metric =metric)

    hc_labels = fcluster(z, criterion = 'maxclust', t = 5)

    hc_silhouette = silhouette_score(df6, hc_labels)
    hc_calinski = calinski_harabasz_score(df6, hc_labels)
    hc_davies = davies_bouldin_score(df6, hc_labels)

    results.append({
        'Method': method,
        'Metric': metric,
        'Silhouette': hc_silhouette,
        'Calinski': hc_calinski,
        'Davies': hc_davies
    })

  except Exception as e:
    print(f"Skipped method = {method}, metric = {metric} due to error {e}")
```

#Pipeline

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
import joblib
```

## Defining Class for feature Engineering

```python
class cxloyaltyfeature(BaseEstimator, TransformerMixin):
  def __init__(self):
    pass

  def fit(self, X, y=None):
    return self

  def transform(self, X):
    data = X.copy()
    data['recency'] = data['recency'].replace(0,1)
    data['cx_loyalty'] = data['customer_since']/data['recency']
    data = data.drop(columns = ['customer_since', 'recency'])
    return data

```

## Preprocess

```python
from sklearn.compose import ColumnTransformer
```

```python
df6.columns
```

```python
numeric_features = ['age', 'family_size', 'income', 'cx_loyalty', 'numdealspurchases',
       'total_amnt', 'total_purchases']
```

```python
preprocessor = ColumnTransformer(
    transformers = [('num', MinMaxScaler(), numeric_features)],
    remainder = 'drop' #passthrough is another option while drop drops other columns
)
```

## Pipeline

```python
kmeans_3 = KMeans(n_clusters = 5, init = 'k-means++', random_state= 42 )
```

```python
clst_pipeline = Pipeline([
    ('cx_loyalty', cxloyaltyfeature()),
    ('scaler', preprocessor),
    ('kmeans', kmeans_3)
])
```
